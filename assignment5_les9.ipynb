```python
import pandas as pd

df = pd.read_excel("/Users/levmanulak/Desktop/ПОДКАСТ/data.xls")

#опис 

df.describe()
```

```python
df_sorted = df.iloc[:1001, :4]

df_sorted.fillna(df_sorted.mean(), inplace=True) 

#тут видно, що результати змінились, 
#тобто ми заповнили поля середнім значенням з перших 1000 рядків, 
#але коли прописую команду df_sorted.isna().any(), то всі рядки мені повертає, як True. Так має бути?
```

```python
df_sorted2 = df.iloc[1001:2001, :4] 

df_sorted2.ffill(inplace=True)
```

```python
df_sorted3 = df.iloc[2001:, :4] 

df_sorted3.bfill(inplace=True)
```

```python
combined_df = pd.concat([df_sorted, df_sorted2, df_sorted3])

combined_df.describe()

#Бачимо, що змінилась к-сть, середнє значення, максимальне і усі відсоткові, але мінімальне залишилось сталим. 
#Можна припустити, що це було дефолтне значення. Єдине, що напружує, то це від'ємне значення min sepal_length. 
#Думаю, це неточні дані, які можна буде очистити в майбутньоиму, але поки такого завдання не було. 
```

```python
import sklearn

X = combined_df[['sepal_length', 'sepal_width', 'petal_length']]
Y = combined_df['petal_width']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train, Y_train)
```

```python
lr.predict(X_test)
pred = lr.predict(X_test)
pred
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

print("Mean absolute error:", mean_absolute_error(Y_test, pred))
print("Mean Squared error:", mean_squared_error(Y_test, pred))

#аби перевірити результати моделі
```
